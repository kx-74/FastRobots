##### Task 2

```python
ble.send_command(CMD.GET_TIME_MILLIS, "")

s = ble.receive_string(ble.uuid['RX_STRING'])
print(s)
```

##### Task 3

```python
def notification_handler(uuid, byte_array):
    print("Received time: " + ble.bytearray_to_string(byte_array))

ble.start_notify(ble.uuid['RX_STRING'], notification_handler)
ble.send_command(CMD.GET_TIME_MILLIS, "")
```

##### Task 4

```python
messages_array = []

def notification_handler(uuid, byte_array):
    print("Received time: " + ble.bytearray_to_string(byte_array))
    messages_array.append(ble.bytearray_to_string(byte_array))

ble.start_notify(ble.uuid['RX_STRING'], notification_handler)
ble.send_command(CMD.GET_TIME_MILLIS_LOOP, "")
```

```python
print("Number of messages received:", len(messages_array))
print("Size of a single message:", len(messages_array[0]))
print("Start time:", messages_array[0], "End time:", messages_array[-1])
```

```python
data_transfer_rate = len(messages_array[0])*len(messages_array) / (0.001*(349646-320692))
print("Effective data transfer rate (byte/s):", data_transfer_rate)
```

##### Task 5

```python
time_array = []

def notification_handler(uuid, byte_array):
    print("Received time: " + ble.bytearray_to_string(byte_array))
    time_array.append(ble.bytearray_to_string(byte_array))

ble.start_notify(ble.uuid['RX_STRING'], notification_handler)
```

```python
ble.send_command(CMD.GET_TIME_MILLIS_LOOP, "")
ble.send_command(CMD.SEND_TIME_DATA, "")
```

```python
print("Number of messages received:", len(time_array))
print("Size of a single message:", len(time_array[0]))
print("Start time:", time_array[1], time_array[-1])
```

```python
data_transfer_rate = len(time_array[0])*len(time_array) / (0.001*(149360-149343))
print("Effective data transfer rate (byte/s):", data_transfer_rate)
```

#### Additional Tasks

```python
ble.send_command(CMD.SEND_MESSAGES, "hey")
```

```python
received_msg = []

def notification_handler(uuid, byte_array):
    print(ble.bytearray_to_string(byte_array))
    received_msg.append(ble.bytearray_to_string(byte_array))

ble.start_notify(ble.uuid['RX_STRING'], notification_handler)
```

##### 5-bytes messages

```python
# 5-bytes messages
received_msg.clear()
ble.send_command(CMD.SEND_MESSAGES, "5-bytes")
```

```python
# Sent the 5-bytes for 1000 times
print(received_msg[0], received_msg[-1], sep='\n')
```

##### 120-bytes messages

```python
received_msg.clear()
ble.send_command(CMD.SEND_MESSAGES, "120-bytes")
```

```python
# Sent the 120-bytes for 400 times
print(received_msg[0], '\n', received_msg[-1], sep='')
```
